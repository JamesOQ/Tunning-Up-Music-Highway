```python
#barplots for crash types

import matplotlib.pyplot as plt
import seaborn as sns

# Count crash types
crash_counts = df['Crash_Type (group)'].value_counts()

# Plot
plt.figure(figsize=(10,7))
crash_counts.plot(kind='barh', color=plt.cm.viridis(range(len(crash_counts))))
plt.title('Crash Type Counts')
plt.xlabel('Number of Crashes')
plt.ylabel('Crash Type')
plt.gca().invert_yaxis()  # Highest count on top
plt.show()

```

```python
#Bar plot: Crash counts by Day of Week

plt.figure(figsize=(10,5))
sns.countplot(x='Collision_DOW', data=df, order=['Monday','Tuesday','Wednesday','Thursday','Friday','Saturday','Sunday'])
plt.title('Number of Crashes by Day of Week')
plt.xlabel('Day of Week')
plt.ylabel('Crash Count')
plt.show()
```

```python
#Scatter Plot: Crash Locations Colored by Severity
plt.figure(figsize=(10, 6))
scatter = plt.scatter(
    df['Longitude_adj'], 
    df['Latitude_adj'], 
    c=df['SUM([Injury]+[Fatality])'], 
    cmap='coolwarm', 
    alpha=0.6
)
plt.colorbar(scatter, label='Injuries + Fatalities')
plt.title('Crash Locations Colored by Severity')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Crash Density Heatmap
plt.figure(figsize=(10, 6))
plt.hist2d(
    df['Longitude_adj'], 
    df['Latitude_adj'], 
    bins=50, 
    cmap='Reds'
)
plt.colorbar(label='Crash Density')
plt.title('Heatmap of Crash Locations')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
 #Histogram: Crashes by Time of Day
# Convert to datetime 
df['Collision_Time_adj'] = pd.to_datetime(df['Collision_Time_adj'], errors='coerce')

# Extract hour
df['Hour'] = df['Collision_Time_adj'].dt.hour

# Plot
plt.figure(figsize=(10, 5))
sns.histplot(df['Hour'].dropna(), bins=24, kde=False, color='#264653')
plt.title('Number of Crashes by Hour of Day')
plt.xlabel('Hour (0â€“23)')
plt.ylabel('Number of Crashes')
plt.xticks(range(0, 24))
plt.tight_layout()
plt.show()
```

```python
#Distribution of Injuries and Fatalities per Crash
import matplotlib.pyplot as plt
import seaborn as sns  

df2['SUM([Injury]+[Fatality])'].value_counts().sort_index().plot(kind='bar', color='orange')
plt.title('Distribution of Injuries and Fatalities per Crash')
plt.xlabel('Number of Injuries and Fatalities')
plt.ylabel('Crash Count')
plt.tight_layout()
plt.show()
 #
```

```python
#Crash Locations Scatter Plot by Latitude and Longitude
plt.figure(figsize=(10,6))
sns.scatterplot(data=df2, x='Longitude_adj', y='Latitude_adj', hue='SUM([Injury]+[Fatality])', palette='coolwarm', alpha=0.6)
plt.title('Crash Locations Colored by Number of Injuries and Fatalities')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.legend(title='Injuries + Fatalities')
plt.tight_layout()
plt.show()
```

```python
#Top Crash Types for Severe and Fatal Crashes
top_crash_types = df2['Crash_Type (group)'].value_counts().head(10)
top_crash_types.plot(kind='bar', color='maroon')
plt.title('Top 10 Crash Types')
plt.xlabel('Crash Type')
plt.ylabel('Count')
plt.tight_layout()
plt.show()
```

```python
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Pavement condition
df2['Pavement condition'].value_counts().plot(kind='bar', ax=axs[0, 0], color='slateblue')
axs[0, 0].set_title('By Pavement Condition')
axs[0, 0].set_ylabel('Crash Count')

# Urban vs Rural
df2['Urban vs. rural environment'].value_counts().plot(kind='bar', ax=axs[0, 1], color='forestgreen')
axs[0, 1].set_title('Urban vs Rural')
axs[0, 1].set_ylabel('Crash Count')

# Nighttime lighting
df2['Nighttime lighting and visibility'].value_counts().plot(kind='bar', ax=axs[1, 0], color='mediumpurple')
axs[1, 0].set_title('Nighttime Lighting & Visibility')
axs[1, 0].set_ylabel('Crash Count')

# Light condition
df2['Light_Condition (group)'].value_counts().plot(kind='bar', ax=axs[1, 1], color='goldenrod')
axs[1, 1].set_title('Lighting Condition')
axs[1, 1].set_ylabel('Crash Count')

plt.tight_layout()
plt.show()
```

```python
# Map numeric values to words for readability
pavement_map = {
    1: 'Dry',
    2: 'Wet',
    3: 'Snow/Slush',
    4: 'Ice',
    5: 'Mud/Dirt',
    6: 'Oily',
    9: 'Unknown'
}
urban_rural_map = {
    1: 'Urban',
    2: 'Rural'
}
lighting_visibility_map = {
    1: 'Well-lit',
    2: 'Dark - No lights',
    3: 'Dark - Lights not working',
    4: 'Twilight/Dawn',
    9: 'Unknown'
}

# Apply the mappings
df2['Pavement condition'] = df2['Pavement condition'].map(pavement_map)
df2['Urban vs. rural environment'] = df2['Urban vs. rural environment'].map(urban_rural_map)
df2['Nighttime lighting and visibility'] = df2['Nighttime lighting and visibility'].map(lighting_visibility_map)

# Create the subplots
fig, axs = plt.subplots(2, 2, figsize=(14, 10))

# Pavement condition
df2['Pavement condition'].value_counts().plot(kind='bar', ax=axs[0, 0], color='slateblue')
axs[0, 0].set_title('By Pavement Condition')
axs[0, 0].set_ylabel('Crash Count')

# Urban vs Rural
df2['Urban vs. rural environment'].value_counts().plot(kind='bar', ax=axs[0, 1], color='forestgreen')
axs[0, 1].set_title('Urban vs Rural')
axs[0, 1].set_ylabel('Crash Count')

# Nighttime lighting
df2['Nighttime lighting and visibility'].value_counts().plot(kind='bar', ax=axs[1, 0], color='mediumpurple')
axs[1, 0].set_title('Nighttime Lighting & Visibility')
axs[1, 0].set_ylabel('Crash Count')

# Light condition
df2['Light_Condition (group)'].value_counts().plot(kind='bar', ax=axs[1, 1], color='goldenrod')
axs[1, 1].set_title('Lighting Condition')
axs[1, 1].set_ylabel('Crash Count')

plt.tight_layout()
plt.show()
```

```python
#Scatter Plot of Crash Locations by Latitude and Longitude, Colored by Crash Severity

import matplotlib.pyplot as plt

plt.figure(figsize=(10, 6))
scatter = plt.scatter(df2['Longitude_adj'], df2['Latitude_adj'], 
                      c=df2['SUM([Injury]+[Fatality])'], cmap='coolwarm', alpha=0.6)
plt.colorbar(scatter, label='Number of Injuries + Fatalities')
plt.xlabel('Longitude')
plt.ylabel('Latitude')
plt.title('Crash Locations on Map Colored by Severity (Injuries + Fatalities)')
plt.show()
```

```python
#Bar Plot of Crash Severity (Number of Injuries + Fatalities)
df2['SUM([Injury]+[Fatality])'].value_counts().sort_index().plot(kind='bar', color='crimson', figsize=(8,5))
plt.title('Distribution of Crash Severity (Injuries + Fatalities)')
plt.xlabel('Number of Injuries + Fatalities')
plt.ylabel('Count of Crashes')
plt.show()
```
